```python
def gradient_at_end_point(last_two_rows, time_step=1, testing = True):
    '''
    last_two_rows : last two rows of table, must be in pd format
    # point : point at which it is observed
    time_step : how fine you want the next step to be
    '''
    
    temp_df = last_two_rows
    (y1_value,y2_value) = temp_df['Adj. Open'].values[0],temp_df['Adj. Open'].values[1] 
    (x1_ind,x2_ind) = temp_df['index'].values[0],temp_df['index'].values[1]
    
    dif_value = y2_value - y1_value
    
    gradient_at_last = float(dif_value)/(x2_ind-x1_ind)
    if testing == True:
        print('per something - add it later')
    
    return gradient_at_last
    
```

```python
df_boring['one_before'] = df_boring['Adj. Open'].shift(1)
```

```python
df_boring['one_before_index'] = df_boring['index'].shift(1)
```

```python
df_boring['gradient_first_order'] = df_boring.apply(lambda x: (x['Adj. Open'] - x['one_before'])/(x['index'] - x['one_before_index']), axis  = 1)
```

```python
df_boring['gradient_first_order_1_before'] = df_boring['gradient_first_order'].shift(1)
```

```python
df_boring['gradient_second_order'] = df_boring.apply(lambda x: (x['gradient_first_order'] - x['gradient_first_order_1_before'])/(x['index'] - x['one_before_index']), axis  = 1)
```
